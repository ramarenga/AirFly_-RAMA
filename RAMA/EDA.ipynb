```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a pandas DataFrame
df = pd.read_csv('/content/flight_delays.csv')

# Example 1: Correlation between numerical features
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Example 2: Relationship between Airline and DelayMinutes
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='Airline', y='DelayMinutes')
plt.title('Airline vs Delay Minutes')
plt.xticks(rotation=45)
plt.show()

# Example 3: Average delay per destination
avg_delay = df.groupby('Destination')['DelayMinutes'].mean().sort_values(ascending=False)
avg_delay.plot(kind='bar', figsize=(10,5), color='teal')
plt.title('Average Delay by Destination')
plt.ylabel('Average Delay (Minutes)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a pandas DataFrame
df = pd.read_csv('/content/flight_delays.csv')

# Boxplot for numerical columns to check outliers
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(data=df, x=col)
    plt.title(f'Outlier Detection - {col}')
    plt.show()

# You can remove outliers using IQR method
Q1 = df['DelayMinutes'].quantile(0.25)
Q3 = df['DelayMinutes'].quantile(0.75)
IQR = Q3 - Q1
df_no_outliers = df[(df['DelayMinutes'] >= (Q1 - 1.5 * IQR)) & (df['DelayMinutes'] <= (Q3 + 1.5 * IQR))]
```

```python
# Check for duplicates
print("Duplicate rows before removal:", df.duplicated().sum())

# Remove duplicate rows
df = df.drop_duplicates()

print("Duplicate rows after removal:", df.duplicated().sum())
```

```python
# Check missing values
df.isnull().sum()

# Example handling:
# For numerical columns → use mean or median
df['DelayMinutes'] = df['DelayMinutes'].fillna(df['DelayMinutes'].median())

# For categorical columns → use mode
df['DelayReason'] = df['DelayReason'].fillna(df['DelayReason'].mode()[0])

# Drop rows if many columns are missing
df = df.dropna(subset=['FlightID', 'Airline'])  # essential columns
```

```python
# Convert columns to datetime
df['ScheduledDeparture'] = pd.to_datetime(df['ScheduledDeparture'], errors='coerce')
df['ActualDeparture'] = pd.to_datetime(df['ActualDeparture'], errors='coerce')
df['ScheduledArrival'] = pd.to_datetime(df['ScheduledArrival'], errors='coerce')
df['ActualArrival'] = pd.to_datetime(df['ActualArrival'], errors='coerce')

# Confirm conversion
df.dtypes
```

```python
# Example: Convert Distance from miles → kilometers (if needed)
if df['Distance'].max() < 1000:  # assume values are in miles
    df['Distance_km'] = df['Distance'] * 1.60934
```

```python
# Create new date/time-based features
df['Year'] = df['ScheduledDeparture'].dt.year
df['Month'] = df['ScheduledDeparture'].dt.month_name()
df['Day'] = df['ScheduledDeparture'].dt.day
df['DayOfWeek'] = df['ScheduledDeparture'].dt.day_name()

# Create Route (Origin → Destination)
df['Route'] = df['Origin'] + ' → ' + df['Destination']

# Create Delay Category
df['DelayCategory'] = pd.cut(df['DelayMinutes'],
                             bins=[-1, 0, 15, 60, 180, np.inf],
                             labels=['No Delay', 'Short Delay', 'Moderate Delay', 'Long Delay', 'Extreme Delay'])
```

```python
# Remove cancelled/diverted flights
df = df[(df['Cancelled'] == 0) & (df['Diverted'] == 0)]

# Remove negative or unrealistic delays
df = df[df['DelayMinutes'] >= 0]

# Optional: Filter out extreme outliers (> 8 hours delay)
df = df[df['DelayMinutes'] <= 480]
```

```python
# Monthly average delay per airline
monthly_delay = df.groupby(['Year', 'Month', 'Airline'])['DelayMinutes'].mean().reset_index()

# Preview
monthly_delay.head()
```
